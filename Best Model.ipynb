```python
# 
# Author: Italo Balestra <italobale@gmail.com>

"""
This script analyses data of the G-Research Financial Forecasting Challenge
and generates predictions based on the best performing model: XGBoost model.
"""

import numpy as np
import pandas as pd
import pickle
from xgboost import XGBRegressor, plot_importance
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import explained_variance_score, mean_absolute_error, mean_squared_error, r2_score 
from sklearn.model_selection import StratifiedKFold, train_test_split, GridSearchCV 
import matplotlib.pyplot as plt

%matplotlib inline

def typlot(pp):
	# Plot Predicted vs True y values
	fig = plt.figure(figsize=(6,5.5))
	plt.plot([-1,1], [-1,1], 'k--', alpha=0.3)
	plt.plot([-1,1], [0,0], 'k:', alpha=0.3)
	plt.plot([0,0], [-1,1], 'k:', alpha=0.3)
	plt.scatter(pp, y_test, alpha=0.75)
	plt.xlim(y_test.min(), y_test.max())
	plt.ylim(y_test.min(), y_test.max())
	plt.ylabel('True y')
	plt.xlabel('Predicted y')
	plt.annotate('MSE:       {0:3.2e}'.format(mean_squared_error(y_test, pp)),
		         xy=(0.7, 0.19), xycoords='figure fraction')
	plt.annotate(r'r$^2$ score: {0:3.5f}'.format(r2_score(y_test, pp)),
		         xy=(0.7, 0.15), xycoords='figure fraction')
	plt.tight_layout()
	plt.show()
	plt.close()
```

```python
# Read Training Dataset

df = pd.read_csv('train.csv', index_col=0)
df.head()
```

```python
# Read Test Dataset

df_test = pd.read_csv('test.csv', index_col=0)
df_test.head()
```

```python
# Quick Look at Statistics of Training Dataset

df.describe().T
```

```python
# Quick Look at Statistics of Test Dataset

df_test.describe().T
```

```python
#####################################################################
#  XGBoost Model
#####################################################################
#
# Data Preparation
#
df = df.dropna()

# Split Dataframe in X and y
X = df.drop(['y','Weight'], axis=1)
y = df['y']

# Split the Data in train and test sets
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)

print('Size of training set: \t{0:d}'.format(X_train.shape[0]))
print('Size of test set: \t{0:d}'.format(X_test.shape[0]))
```

```python
# Define the model
model = XGBRegressor(n_estimators=300, max_depth=7, colsample_bytree = 0.8)

# Fit model on training data
#
# Define an evaluation subset to monitor performance and set early stopping 
eval_set = [(X_test, y_test)]

model.fit(X_train, y_train, eval_metric="logloss", eval_set=eval_set, verbose=False)

#[299]	validation_0-logloss:-0.00132
```

```python
# Make predictions for test data
predictions = model.predict(X_test)

# Plot comparison of true values with model predictions
typlot(predictions)
```

```python
# Print scores
print('MSE:\t', mean_squared_error(y_test, predictions))
print('MAE:\t', mean_absolute_error(y_test, predictions))
print('r^2 score:\t', r2_score(y_test, predictions)) 
print('Expl. Var.:\t', explained_variance_score(y_test, predictions)) # 1.0 is Good, lower are bad
```

```python
# Feature Importance 
for f, fimp in enumerate(model.feature_importances_):
	print('Feature '+str(f)+': '+str(fimp))
```

```python
plot_importance(model)
plt.show()
plt.close()
```

```python
#####################################################################
#  Run XGBoost Model on the Challenge Test Dataset 
#####################################################################
#

# Split Dataframe in X and y
X_train = df.drop(['y','Weight'], axis=1)
y_train = df['y']
X_test = df_test

# Fit model on training data
model = XGBRegressor(n_estimators=560, max_depth=7, colsample_bytree = 0.8)

model.fit(X_train, y_train, eval_metric="logloss", eval_set=eval_set, verbose=False)

#[559]	validation_0-logloss:-0.002102
```

```python
# Make predictions for test data
y_pred = model.predict(X_test)

########################################################
#  PREPARE RESULTS ON A csv FILE TO SUBMIT THEM
########################################################

# Write Data to csv file
yp = pd.Series(y_pred).rename('y')
yp.index.name = 'Index'
yp.to_csv('XGBoost_14_model.csv', header=True)

# Save model to file
pickle.dump(model, open('XGBoost_14_model.pickle.dat', 'wb'))
```
